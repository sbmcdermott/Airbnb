```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
results = pd.read_csv('data_for_viz/weekdays.csv')
results.drop(columns = ['Unnamed: 0'], axis=1, inplace=True)
results.head()
```

```python
results.head(25)
```

```python
# Creating dataset 
a = results['guests']

# Creating histogram 
fig, ax = plt.subplots(figsize =(10, 7)) 
ax.hist(a , bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85) 
plt.title('Listings Per Number of Guests')  

# Show plot 
plt.show() 
```

```python
plt.style.use('seaborn-whitegrid')
x = results['guests']
y = results['total cost']

plt.xlabel('Number of Guests')
plt.ylabel('Total Cost')

plt.plot(x, y, 'o', color='blue', alpha=0.1);
```

```python
checkin_grouped = results.groupby('checkin')
grouped_mean = checkin_grouped.mean()
grouped_mean.reset_index(inplace=True)
grouped_mean.head()
```

```python
plt.plot(grouped_mean['checkin'], grouped_mean['total cost'], color='blue', marker='o')
plt.title('Average Price Per Check In Date', fontsize=14)
plt.xlabel('Check In Date', fontsize=14)
plt.ylabel('Average Price Per Listing', fontsize=14)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
x = results['avg_stars']
y = results['review_count']

plt.xlabel('Star Rating')
plt.ylabel('Number of Reviews')

plt.plot(x, y, 'o', color='blue', alpha=0.1);
```

```python
guests_grouped = results.groupby('guests')
guests_mean = guests_grouped.mean()
guests_mean.reset_index(inplace=True)
guests_mean.head(20)
```

```python
# Creating bar chart
plt.bar(guests_mean['guests'], guests_mean['review_count'])
plt.title('Reviews per Guest Count')  

# Show plot 
plt.show() 
```

```python
# Creating bar chart
plt.bar(guests_mean['guests'], guests_mean['total cost'])
plt.title('Total Cost v Guests')  

# Show plot 
plt.show() 
```

```python
# Creating bar chart
plt.bar(guests_mean['guests'], guests_mean['room_fee'])
plt.title('Total Room Fee v Guests')  

# Show plot 
plt.show()
```

```python
# Creating bar chart
plt.bar(guests_mean['guests'], guests_mean['current_nightly'])
plt.title('Nightly Rate v Guests')  

# Show plot 
plt.show()
```

```python
# Creating bar chart
plt.bar(guests_mean['guests'], guests_mean['avg_stars'])
plt.title('Rating v Guests')  

# Show plot 
plt.show()
```

```python
pricedrop = results[results['previous_price'].notna()]
pricedrop.head()
```

```python
pricedrop['previous_price'] = pricedrop['previous_price'].astype(int)
pricedrop['discounted'] = pricedrop['discounted'].astype(int)
```

```python
pricedrop['price_reduc'] = pricedrop['previous_price'] - pricedrop['discounted']
pricedrop.head()
```

```python
pricedrop['perc_reduc'] = pricedrop['price_reduc'] / pricedrop['previous_price']
pricedrop.head()
```

```python
# Creating bar chart
plt.bar(pricedrop['guests'], pricedrop['perc_reduc'])
plt.title('Price Drop v Guests')  

# Show plot 
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
x = pricedrop['total cost']
y = pricedrop['perc_reduc']

plt.xlabel('Total Cost')
plt.ylabel('perc_reduc')

plt.plot(x, y, 'o', color='blue', alpha=0.1);
```

```python
plt.style.use('seaborn-whitegrid')
x = pricedrop['total cost']
y = pricedrop['taxes_fees_cleaning']

plt.xlabel('Total Cost')
plt.ylabel('perc_reduc')

plt.plot(x, y, 'o', color='blue', alpha=0.1);
```

```python
plt.style.use('seaborn-whitegrid')
x = pricedrop['guests']
y = pricedrop['taxes_fees_cleaning']

plt.xlabel('Total Cost')
plt.ylabel('perc_reduc')

plt.plot(x, y, 'o', color='blue', alpha=0.1);
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Price Reduction (%)')
ax1.boxplot(pricedrop['perc_reduc']);
```

```python
# Creating dataset 
a = results['category'].value_counts()[:10]
a = pd.DataFrame(a).reset_index()

# Creating bar chart
plt.bar(a['index'], a['category'])
plt.title('Type of Listings')  
plt.xticks(rotation=90)

# Show plot 
plt.show()
```
