```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
results = pd.read_csv('data_for_viz/weekends.csv')
results.drop(columns = ['Unnamed: 0'], axis=1, inplace=True)
results.head()
```

```python
# Creating dataset 
a = results['guests']

# Creating histogram 
fig, ax = plt.subplots(figsize =(10, 7)) 
ax.hist(a , bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85) 
plt.title('Listings Per Number of Guests')  

# Show plot 
plt.show() 
```
